## Plot the following function with appropriate labels and titles

	
$\huge y=e ^{-x} sin(2x+\frac{\pi}{3})$ 


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-2*np.pi, 2*np.pi, 0.1)
y = np.exp(-x)*np.sin(2*x+np.pi/3)


plt.plot(x,y + 2, '-.g')
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.grid(True)
plt.title(r"$y=e ^{-x} sin(2x+\frac{\pi}{3})$", fontsize=18);
plt.show()
```

## Generate the term n of a recursive relationship given by
$\largeùëéùëõ=5ùëéùëõ‚àí1+2,  ùëé0=2$

```python
def sequence(n):
    # define an empty list for the sequence
    sequence_list = []
    # define first element value
    last_element = 2
    # append the list with first element
    sequence_list.append(last_element)
    # loop n-1 times because first element is appended in the list
    for i in range(n-1):
        # calculate next element based on sequence
        current_element = 5 * last_element + 2
        # update last element with current element
        last_element = current_element
        # append the current element in the list
        sequence_list.append(current_element)
        
    # return the list
    return sequence_list

def main():
    # validate user input
    while True:
        try:
            # convert the user input into integer
            n = int(input("Enter N Val:"))
            if n <= 0:
                # in case user enter value less than or equal to zero the program will raise exception
                raise Exception
                
            # in case of valid input calculate sequence with sequence function
            sequence_list = sequence(n)
            print(sequence_list)
            # break the loop
            break
        except:
            # in case of invalid input print the message
            print("Invalid Input")


main()
```

## Develop a python script to find the 3-digit pin number in the previous task. Assume that we do not know the pin number and the script should find be able to restore the pin.


```python
from random import randint 

def pin_generator():
    # select a random three digit number and return it
    value = randint(100, 999)
    return value
    
def main():
    # get the pin from pin generator function
    pin = pin_generator()
    # loop through each three digit number
    for i in range(100,1000):
        # check if current i value match the pin
        if i == pin:
            # if value matches print the pin and break the loop
            print("This is your 3-digit number", pin)
            break
            

main()
```
